```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import Audio

# Define parameters
duration = 5  # seconds
fs = 44100  # sampling frequency in Hz
frequencies = [500, 5000, 10000]  # Hz

# Generate time vector
t = np.linspace(0, duration, int(fs * duration), endpoint=False)

signals = {}

# Generate signals for each frequency
for f in frequencies:
    x_t = np.cos(2 * np.pi * f * t)
    signals[f] = x_t
    print(f"Generated signal for f = {f} Hz. Length: {len(x_t)} samples.")
```

### (a) Gerar gráficos de x(t) no domínio do tempo para diferentes frequências.

```python
plt.figure(figsize=(15, 10))

for i, f in enumerate(frequencies):
    plt.subplot(len(frequencies), 1, i + 1)
    # Plot only a small fraction of the signal to visualize the waveform
    plt.plot(t[:int(fs * 0.01)], signals[f][:int(fs * 0.01)]) # Plot first 0.01 seconds
    plt.title(f'Sinal x(t) para f = {f} Hz (Primeiros 0.01s)')
    plt.xlabel('Tempo (s)')
    plt.ylabel('Amplitude')
    plt.grid(True)

plt.tight_layout()
plt.show()
```

### (b) Escreva um código para tocar os sinais gerados.

```python
for f in frequencies:
    print(f"Reproduzindo sinal para f = {f} Hz:")
    # Normalize the signal to prevent clipping when playing
    audio_data = signals[f] / np.max(np.abs(signals[f]))
    display(Audio(audio_data, rate=fs))
    print("\n" * 2) # Add some space between audio players
```

### (c) Comente os resultados obtidos.

**Observações sobre os Gráficos de Tempo:**

*   **Frequência de 500 Hz:** O gráfico mostra um sinal de cosseno que oscila relativamente devagar. É possível observar várias ondas completas no período de 0.01 segundos, o que é esperado para uma frequência baixa.
*   **Frequência de 5000 Hz:** O sinal oscila muito mais rapidamente do que o de 500 Hz. No mesmo período de 0.01 segundos, há muito mais ciclos da onda, indicando uma frequência mais alta.
*   **Frequência de 10000 Hz:** Este sinal apresenta a maior taxa de oscilação entre os três. No período de 0.01 segundos, as oscilações são muito densas, o que é característico de uma alta frequência.

Todos os sinais são senoidais puros, com amplitude constante, como esperado de um sinal `cos(2πf t)`.

**Observações sobre a Reprodução dos Sinais:**

*   **Frequência de 500 Hz:** Ao reproduzir, este sinal deve soar como um tom grave, contínuo e suave. É uma frequência que está bem dentro do alcance da audição humana e é percebida claramente como um som musical.
*   **Frequência de 5000 Hz:** Este sinal deve soar como um tom agudo, também contínuo. É significativamente mais alto em tom do que o sinal de 500 Hz.
*   **Frequência de 10000 Hz:** O sinal de 10000 Hz é um tom muito agudo. Algumas pessoas podem ter dificuldade em ouvir frequências tão altas com a mesma clareza que as frequências mais baixas, dependendo da sua audição e da qualidade dos dispositivos de áudio.

Em geral, a experiência auditiva confirma a teoria de que frequências mais baixas correspondem a tons graves e frequências mais altas a tons agudos.
